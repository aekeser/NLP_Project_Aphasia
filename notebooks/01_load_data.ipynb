```python
def extract_cat_features(file_path):
    import pylangacq
    
    chat = pylangacq.read_chat(file_path)
    utts = chat.utterances()
    
    # find Cat segment
    start = None
    end = None
    
    for i, utt in enumerate(utts):
        if "Cat" in str(utt):
            start = i + 1
        elif start is not None and "G {" in str(utt):
            end = i
            break
    
    if start is None or end is None:
        return None
    
    segment = utts[start:end]
    
    # keep PAR utterances
    par_utts = [utt for utt in segment if "participant='PAR'" in str(utt)]
    
    if len(par_utts) == 0:
        return None
    
    # collect words
    words = []
    for utt in par_utts:
        for tok in utt.tokens:
            w = tok.word.lower()
            if w.isalpha():
                words.append(w)
    
    if len(words) == 0:
        return None
    
    # MLU
    mlu = len(words) / len(par_utts)
    
    # TTR
    ttr = len(set(words)) / len(words)
    
    # function-word ratio
    function_words = {
        'the','a','an','and','or','but','to','of','in','on','at','for',
        'with','by','from','is','are','was','were','be','been','being',
        'he','she','it','they','him','her','them',
        'his','hers','their',
        'that','this','these','those',
        'who','what','where','when','why','how'
    }
    
    func_count = sum(1 for w in words if w in function_words)
    func_ratio = func_count / len(words)
    
    return mlu, ttr, func_ratio
```

```python
def get_task_segments(utts):
    segments = []
    current_task = None
    start_idx = None

    for i, utt in enumerate(utts):
        utt_str = str(utt)

        if "G { value: Some(" in utt_str:
            # save previous segment
            if current_task is not None:
                segments.append((current_task, start_idx, i))

            # extract task name
            task_name = utt_str.split('Some("')[1].split('")')[0]
            current_task = task_name
            start_idx = i + 1

    # last segment
    if current_task is not None:
        segments.append((current_task, start_idx, len(utts)))

    return segments
```

```python
def extract_task_features(file_path, task_name):
    import pylangacq

    chat = pylangacq.read_chat(file_path)
    utts = chat.utterances()
    segments = get_task_segments(utts)

    matching_segments = [
        (task, start, end)
        for task, start, end in segments
        if task.lower() == task_name.lower()
    ]

    if len(matching_segments) == 0:
        return None

    task, start, end = matching_segments[0]
    segment = utts[start:end]

    par_utts = [utt for utt in segment if "participant='PAR'" in str(utt)]

    if len(par_utts) == 0:
        return None

    words = []
    for utt in par_utts:
        for tok in utt.tokens:
            w = tok.word.lower()
            if w.isalpha():
                words.append(w)

    if len(words) == 0:
        return None

    mlu = len(words) / len(par_utts)
    ttr = len(set(words)) / len(words)

    function_words = {
        'the','a','an','and','or','but','to','of','in','on','at','for',
        'with','by','from','is','are','was','were','be','been','being',
        'he','she','it','they','him','her','them',
        'his','hers','their',
        'that','this','these','those',
        'who','what','where','when','why','how'
    }

    func_count = sum(1 for w in words if w in function_words)
    func_ratio = func_count / len(words)

    return mlu, ttr, func_ratio
```

```python
from pathlib import Path
import pandas as pd

raw_dir = Path("../data/raw")

corpora = [
    "ACWT",
    "Adler",
    "Baycrest",
    "BU",
    "Fridriksson",
    "Fridriksson_2",
    "Garrett",
    "Kansas",
    "Kempler",
    "Kurland",
    "MCU",
    "Richardson",
    "SCALE",
    "Thompson",
    "Tucson",
    "UMD",
    "UNH",
    "Whiteside",
    "Williamson",
    "Wozniak",
    "Wright"
]

audit_rows = []

for corpus in corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha")) if pwa_dir.exists() else []

    audit_rows.append({
        "corpus": corpus,
        "pwa_folder_exists": pwa_dir.exists(),
        "n_cha_files": len(cha_files)
    })

corpus_audit_df = pd.DataFrame(audit_rows)

corpus_audit_df
```

```python
def extract_aq(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            if line.startswith("@ID:") and "|PAR|" in line:
                parts = line.strip().split("|")

                for value in reversed(parts):
                    value = value.strip()
                    if value == "":
                        continue

                    try:
                        return float(value)
                    except ValueError:
                        continue

                return None

    return None
```

```python
aq_audit_rows = []

for corpus in corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    total_files = 0
    aq_found = 0
    aq_missing = 0

    for file_path in cha_files:
        total_files += 1
        aq = extract_aq(str(file_path))

        if aq is not None:
            aq_found += 1
        else:
            aq_missing += 1

    aq_audit_rows.append({
        "corpus": corpus,
        "total_files": total_files,
        "aq_found": aq_found,
        "aq_missing": aq_missing
    })

aq_audit_df = pd.DataFrame(aq_audit_rows)

aq_audit_df
```

```python
aq_values_rows = []

for corpus in corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    for file_path in cha_files:
        aq = extract_aq(str(file_path))

        if aq is not None:
            aq_values_rows.append({
                "corpus": corpus,
                "file": file_path.name,
                "aq": aq
            })

aq_values_df = pd.DataFrame(aq_values_rows)

aq_range_df = aq_values_df.groupby("corpus")["aq"].agg(
    n="count",
    min="min",
    max="max",
    mean="mean"
).round(2).reset_index()

aq_range_df
```

```python
from collections import Counter

task_counter = Counter()
corpus_task_counter = {}

for corpus in corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    corpus_task_counter[corpus] = Counter()

    for file_path in cha_files:
        try:
            chat = pylangacq.read_chat(str(file_path))
            segments = get_task_segments(chat.utterances())

            for task, start, end in segments:
                task_counter[task] += 1
                corpus_task_counter[corpus][task] += 1

        except Exception:
            continue

# overall task counts
task_df = pd.DataFrame(task_counter.items(), columns=["task", "count"]).sort_values(by="count", ascending=False)

task_df.head(20)
```

```python
target_tasks = ["Cat", "Umbrella", "Window"]

rows = []

for corpus in corpora:
    task_counts = corpus_task_counter[corpus]

    for task in target_tasks:
        rows.append({
            "corpus": corpus,
            "task": task,
            "count": task_counts.get(task, 0)
        })

task_coverage_df = pd.DataFrame(rows)

task_coverage_pivot = task_coverage_df.pivot(index="corpus", columns="task", values="count").fillna(0).astype(int)

task_coverage_pivot
```

```python
def extract_task_features(file_path, task_name):
    import pylangacq

    try:
        chat = pylangacq.read_chat(file_path)
    except Exception:
        try:
            chat = pylangacq.read_chat(file_path, strict=False)
        except Exception:
            return None

    utts = chat.utterances()
    segments = get_task_segments(utts)

    matching_segments = [
        (task, start, end)
        for task, start, end in segments
        if task.lower() == task_name.lower()
    ]

    if len(matching_segments) == 0:
        return None

    task, start, end = matching_segments[0]
    segment = utts[start:end]

    par_utts = [utt for utt in segment if "participant='PAR'" in str(utt)]

    if len(par_utts) == 0:
        return None

    words = []

    for utt in par_utts:
        for tok in utt.tokens:
            w = tok.word.lower()
            if w.isalpha():
                words.append(w)

    if len(words) == 0:
        return None

    mlu = len(words) / len(par_utts)
    ttr = len(set(words)) / len(words)

    function_words = {
        'the','a','an','and','or','but','to','of','in','on','at','for',
        'with','by','from','is','are','was','were','be','been','being',
        'he','she','it','they','him','her','them',
        'his','hers','their',
        'that','this','these','those',
        'who','what','where','when','why','how'
    }

    func_count = sum(1 for w in words if w in function_words)
    func_ratio = func_count / len(words)

    return mlu, ttr, func_ratio
```

```python
rows = []

for corpus in corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    for file_path in cha_files:
        aq = extract_aq(str(file_path))
        if aq is None:
            continue

        for task in ["Cat", "Umbrella", "Window"]:
            features = extract_task_features(str(file_path), task)

            if features is not None:
                rows.append({
                    "corpus": corpus,
                    "file": file_path.name,
                    "task": task
                })

usable_df = pd.DataFrame(rows)

usable_counts = usable_df.groupby("task").agg(
    n_obs=("file", "count"),
    n_participants=("file", "nunique")
)

usable_counts
```

```python
filtered_corpora = [
    corpus for corpus in corpora
    if corpus != "Fridriksson_2"
]

filtered_corpora
```

```python
rows = []

for corpus in filtered_corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    for file_path in cha_files:
        aq = extract_aq(str(file_path))
        if aq is None:
            continue

        for task in ["Cat", "Umbrella", "Window"]:
            features = extract_task_features(str(file_path), task)

            if features is not None:
                rows.append({
                    "corpus": corpus,
                    "file": file_path.name,
                    "task": task
                })

usable_df_filtered = pd.DataFrame(rows)

usable_counts_filtered = usable_df_filtered.groupby("task").agg(
    n_obs=("file", "count"),
    n_participants=("file", "nunique")
)

usable_counts_filtered
```

```python
task_participants = {}

for task in ["Cat", "Umbrella", "Window"]:
    task_participants[task] = set(
        usable_df_filtered[usable_df_filtered["task"] == task]["file"]
    )

common_participants = set.intersection(*task_participants.values())

len(common_participants)
```

```python
rows = []

for corpus in filtered_corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    for file_path in cha_files:
        file_name = file_path.name

        if file_name not in common_participants:
            continue

        aq = extract_aq(str(file_path))
        if aq is None:
            continue

        for task in ["Cat", "Umbrella", "Window"]:
            features = extract_task_features(str(file_path), task)

            if features is not None:
                mlu, ttr, func_ratio = features
                rows.append({
                    "corpus": corpus,
                    "file": file_name,
                    "task": task,
                    "aq": aq,
                    "mlu": mlu,
                    "ttr": ttr,
                    "function_word_ratio": func_ratio
                })

final_df = pd.DataFrame(rows)

final_df.shape
```

```python
final_df["severity"] = final_df["aq"].apply(lambda x: 0 if x < 75 else 1)

final_df["severity"].value_counts()
```

```python
from sklearn.model_selection import GroupShuffleSplit

X = final_df[["mlu", "ttr", "function_word_ratio"]]
y = final_df["severity"]
groups = final_df["file"]

gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42
)

train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

len(train_idx), len(test_idx), final_df.iloc[test_idx]["file"].nunique()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(accuracy)
print(classification_report(y_test, y_pred))
```

```python
coef_df = pd.DataFrame({
    "feature": ["mlu", "ttr", "function_word_ratio"],
    "coefficient": model.coef_[0]
}).sort_values(by="coefficient", ascending=False)

coef_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000, random_state=42))
])

pipeline.fit(X_train, y_train)

y_pred_scaled = pipeline.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(y_test, y_pred_scaled))
print(classification_report(y_test, y_pred_scaled))

coef_scaled = pd.DataFrame({
    "feature": ["mlu", "ttr", "function_word_ratio"],
    "coefficient": pipeline.named_steps["logreg"].coef_[0]
}).sort_values(by="coefficient", ascending=False)

coef_scaled
```

```python
def extract_task_features(file_path, task_name):
    import pylangacq

    try:
        chat = pylangacq.read_chat(file_path)
    except Exception:
        try:
            chat = pylangacq.read_chat(file_path, strict=False)
        except Exception:
            return None

    utts = chat.utterances()
    segments = get_task_segments(utts)

    matching_segments = [
        (task, start, end)
        for task, start, end in segments
        if task.lower() == task_name.lower()
    ]

    if len(matching_segments) == 0:
        return None

    task, start, end = matching_segments[0]
    segment = utts[start:end]

    par_utts_with_words = []
    words = []

    for utt in segment:
        if "participant='PAR'" not in str(utt):
            continue

        utt_words = []

        for tok in utt.tokens:
            w = tok.word.lower()
            if w.isalpha():
                utt_words.append(w)

        if len(utt_words) > 0:
            par_utts_with_words.append(utt)
            words.extend(utt_words)

    if len(par_utts_with_words) == 0 or len(words) == 0:
        return None

    mlu = len(words) / len(par_utts_with_words)
    ttr = len(set(words)) / len(words)

    function_words = {
        'the','a','an','and','or','but','to','of','in','on','at','for',
        'with','by','from','is','are','was','were','be','been','being',
        'he','she','it','they','him','her','them',
        'his','hers','their',
        'that','this','these','those',
        'who','what','where','when','why','how'
    }

    func_count = sum(1 for w in words if w in function_words)
    func_ratio = func_count / len(words)

    return mlu, ttr, func_ratio
```

```python
rows = []

for corpus in filtered_corpora:
    pwa_dir = raw_dir / corpus / "PWA"
    cha_files = sorted(pwa_dir.glob("*.cha"))

    for file_path in cha_files:
        file_name = file_path.name

        if file_name not in common_participants:
            continue

        aq = extract_aq(str(file_path))
        if aq is None:
            continue

        for task in ["Cat", "Umbrella", "Window"]:
            features = extract_task_features(str(file_path), task)

            if features is not None:
                mlu, ttr, func_ratio = features
                rows.append({
                    "corpus": corpus,
                    "file": file_name,
                    "task": task,
                    "aq": aq,
                    "mlu": mlu,
                    "ttr": ttr,
                    "function_word_ratio": func_ratio
                })

final_df = pd.DataFrame(rows)
final_df["severity"] = final_df["aq"].apply(lambda x: 0 if x < 75 else 1)

final_df.shape
```
